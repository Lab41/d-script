```python
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt
from sklearn import manifold
from sklearn.utils import check_random_state
import os

%pylab notebook
```

```python
#corpdir = '/fileserver/nmec-handwriting/stil-writing-corpus-processed/TASfeats/FR'
corpdir = '/fileserver/icdar13/jimbo/processed/TASfeats/'
files = os.listdir(corpdir)
files.sort()

feature_map = []
for filename in files:
    path = corpdir + "/" + filename
    if os.path.isfile(path):
        feature_file = open(path, 'r').read().splitlines()
        feature_map += ([np.fromstring(line, dtype=np.float32, sep=' ') for line in feature_file])
```

```python
feature_map = np.array(feature_map)
feature_map.shape
```

```python
TSNE = manifold.TSNE(learning_rate=100).fit_transform(feature_map).T
```

```python
colors = np.array([x / 4 for x in range(len(feature_map))])
scatter(TSNE[0], TSNE[1], cmap='jet', c=colors)
```
